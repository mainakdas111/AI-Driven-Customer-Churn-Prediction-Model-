### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

## Part 1 - Data Preprocessing

### Importing the dataset

```python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
# Summary statistics for numerical features
numerical_features = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
print(dataset[numerical_features].describe())
```

```python
# Distribution plots for numerical features
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=dataset, x=feature, bins=30, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Frequency distribution for categorical features
categorical_features = ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember', 'Exited']
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=dataset, x=feature, hue=feature)
    plt.title(f'Frequency distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
# Churn rates by categorical features
for feature in categorical_features[:-1]:  # Exclude 'Exited' since it's the target variable
    plt.figure(figsize=(8, 6))
    sns.barplot(data=dataset, x=feature, y='Exited', estimator=lambda x: sum(x) / len(x), hue=feature)
    plt.title(f'Churn rates by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Churn Rate')
    plt.show()
```

```python
# Box plots for numerical features by churn status
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=dataset, x='Exited', y=feature)
    plt.title(f'{feature} by Churn Status')
    plt.xlabel('Churn Status')
    plt.ylabel(feature)
    plt.xticks([0, 1], ['Not Churned', 'Churned'])
    plt.show()
```

### Encoding categorical data

Label Encoding the "Gender" column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
```

```python
print(X)
```

One Hot Encoding the "Geography" column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Apply One Hot Encoding to the column at index 1 of the dataset X
# 'remainder' parameter specifies what to do with remaining columns - in this case, leave them as is ('passthrough')
column_transformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')

# Fit and transform the data, converting the result back into a numpy array as the transformed data is returned as a sparse matrix
X = np.array(column_transformer.fit_transform(X))
```

```python
print(X)
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Part 2 - Building the ANN

### Initializing the ANN

```python
ann = Sequential()
```

### Adding the input layer and the first hidden layer

```python
ann.add(Dense(units=6, activation='relu'))
```

### Adding the second hidden layer

```python
ann.add(Dense(units=6, activation='relu'))
```

**Adding the third hidden layer**

```python
ann.add(Dense(units=3, activation='relu'))
```

### Adding the output layer

```python
ann.add(Dense(units=1, activation='sigmoid'))
```

## Part 3 - Training the ANN

### Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Training the ANN on the Training set

```python
history = ann.fit(X_train, y_train, batch_size=32, epochs=100, validation_split=0.2)
```

```python
# Plotting training history for accuracy
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

## Part 4 - Making the predictions and evaluating the model

### Predicting the result of a single observation

Use our ANN model to predict if the customer with the following informations will leave the bank:

Geography: France

Credit Score: 600

Gender: Male

Age: 40 years old

Tenure: 3 years

Balance: \$ 60000

Number of Products: 2

Does this customer have a credit card ? Yes

Is this customer an Active Member: Yes

Estimated Salary: \$ 50000

So, should we say goodbye to that customer ?

**Solution**

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
```

Therefore, our ANN model predicts that this customer stays in the bank!

**Important note 1:** Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting our values into a double pair of square brackets makes the input exactly a 2D array.

**Important note 2:** Notice also that the "France" country was not input as a string in the last column but as "1, 0, 0" in the first three columns. That's because of course the predict method expects the one-hot-encoded values of the state, and as we see in the first row of the matrix of features X, "France" was encoded as "1, 0, 0". And be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

### Predicting the Test set results

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
